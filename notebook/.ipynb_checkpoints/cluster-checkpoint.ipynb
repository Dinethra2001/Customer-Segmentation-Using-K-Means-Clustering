<h1>Import the DataSet</h1>

```python
# Import required library
import pandas as pd

# Replace 'Online Retail.csv' with the correct filename or path
df = pd.read_csv('data/data.csv', encoding='ISO-8859-1')

# Display the first few rows to confirm it's loaded
df.head()
```

<h1>Preprocessing</h1>

```python
# Check for any missing values in each column
print("Missing values per column:\n", df.isnull().sum())

# Specifically check how many rows have missing CustomerID
missing_customer_ids = df['CustomerID'].isnull().sum()
print(f"\nNumber of rows with missing CustomerID: {missing_customer_ids}")
```

```python
# Count how many rows have negative quantity
negative_quantity_count = (df['Quantity'] < 0).sum()
print(f"Number of rows with negative Quantity: {negative_quantity_count}")
```

```python
# 1. Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# 2. Remove rows with negative or zero Quantity or UnitPrice
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# 3. Calculate TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# 4. Optional: Reset index after filtering
df.reset_index(drop=True, inplace=True)

# Preview cleaned data
df.head()
```

```python
print(f"Remaining rows after cleaning: {df.shape[0]}")
print(f"Columns: {df.columns.tolist()}")
```

<h1>Feature Engineering</h1>

```python
# Convert InvoiceDate to datetime if not already
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Group and aggregate
customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',         # Frequency
    'Quantity': 'sum',              # TotalUnits
    'TotalPrice': 'sum',            # TotalSpent
    'InvoiceDate': ['min', 'max']   # For recency calculation
}).reset_index()
```

```python
# Fix column names (multi-index from aggregation)
customer_df.columns = ['CustomerID', 'Frequency', 'TotalUnits', 'TotalSpent', 'FirstPurchase', 'LastPurchase']
```

```python
# Set reference date as the last date in the dataset
reference_date = df['InvoiceDate'].max()

# Recency = days since last purchase
customer_df['Recency'] = (reference_date - customer_df['LastPurchase']).dt.days
```

```python
customer_df.drop(['FirstPurchase', 'LastPurchase'], axis=1, inplace=True)
```

```python
# Preview the customer-level dataset
customer_df.head()
```

<h1>Feature Scaling</h1>

```python
from sklearn.preprocessing import StandardScaler
X = customer_df[['Recency', 'Frequency', 'TotalUnits', 'TotalSpent']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Convert to DataFrame for readability
scaled_df = pd.DataFrame(X_scaled, columns=['Recency', 'Frequency', 'TotalUnits', 'TotalSpent'])
scaled_df.head()
```

<h1>Dimensionality Reduction (PCA)</h1>

```python
from sklearn.decomposition import PCA
# Initialize PCA to reduce to 2 components
pca = PCA(n_components=2)

# Fit and transform the scaled features
X_pca = pca.fit_transform(X_scaled)
```

```python
# Show how much variance each principal component explains
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained by 2 components:", pca.explained_variance_ratio_.sum())

# Convert PCA result into a DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df.head()
```

<h1>Clustering (K-Means)</h1>

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Store inertia values for different k values
inertia = []

# Test k values from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
# Fit K-Means with chosen number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Add PCA components to the customer_df
customer_df['PCA1'] = X_pca[:, 0]
customer_df['PCA2'] = X_pca[:, 1]

# Preview with cluster and PCA data
customer_df.head()
```

<h1>Visualization & Insights</h1>

Scatter Plot of Clusters in PCA-Reduced 2D Space

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    customer_df['PCA1'], customer_df['PCA2'],
    c=customer_df['Cluster'], cmap='viridis', alpha=0.7
)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Customer Segments (PCA Reduced Space)')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()
```

Pair Plot of Original Features (Color by Cluster)

```python
import seaborn as sns

sns.pairplot(
    customer_df,
    vars=['Recency', 'Frequency', 'TotalSpent'],
    hue='Cluster',
    palette='husl',
    plot_kws={'alpha': 0.6}
)
plt.suptitle('Pairplot of Behavioral Features by Cluster', y=1.02)
plt.show()
```

Boxplot to Compare Features Across Clusters

```python
# Melt the data for seaborn boxplot
melted = pd.melt(
    customer_df,
    id_vars=['Cluster'],
    value_vars=['Recency', 'Frequency', 'TotalUnits', 'TotalSpent']
)

plt.figure(figsize=(12, 6))
sns.boxplot(x='variable', y='value', hue='Cluster', data=melted)
plt.title("Distribution of Features by Cluster")
plt.grid(True)
plt.show()
```

```python
# Mean of each feature by cluster
cluster_summary = customer_df.groupby('Cluster')[['Recency', 'Frequency', 'TotalUnits', 'TotalSpent']].mean()
print(cluster_summary)
```
